```python
import numpy as np
import numpy.random as rn
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['legend.fontsize'] = 8
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.serif'] = ['Computer Modern']
matplotlib.rc('xtick', labelsize=8) 
matplotlib.rc('ytick', labelsize=8)

lightblue = '#79add2'
blue = '#1f77b4'
darkblue = '#13476c'
```

#### Steady state ####

```python
var_T_osys_prm6 = np.load("./data/steady_state/var_T_osys_prm6.npz")
var_T_sto_prm6 = np.load("./data/steady_state/var_T_sto_prm6_20000.npz")
var_T_osys_prm7 = np.load("./data/steady_state/var_T_osys_prm7.npz")
```

#### Dynamics ####

```python
### Import data ###
var_t_osys_prm3 = np.load("./data/dynamics/var_t_osys_prm3.npz")
var_t_osys_prm3_mm = np.array(pd.read_csv(r'./data/dynamics/var_t_osys_prm3_mm.csv', header=None))
var_t_osys_prm3_cl = np.load("./data/dynamics/var_t_osys_prm3_cl.npz")

var_t_sto_prm3 = np.load("./data/dynamics/var_t_sto_prm3.npz")
var_t_osys_prm3_cl_mm = np.array(pd.read_csv(r'./data/dynamics/var_t_osys_prm3_cl_mm.csv', header=None))
var_t_sto_prm3_cl = np.load("./data/dynamics/var_t_sto_prm3_cl.npz")
```

```python
fig = plt.figure(figsize=(3, 5))
gs = fig.add_gridspec(16, 1, hspace=2)
ax1 = fig.add_subplot(gs[0:3, :])
ax2 = fig.add_subplot(gs[3:6, :])
ax3 = fig.add_subplot(gs[6:9, :])
ax4 = fig.add_subplot(gs[9:16, :])

ax1.set_xlim(0, 8)
ax1.set_ylim(0.45, 1.32)
ax2.set_xlim(0, 8)
ax2.set_ylim(-0.37, 0.37)
ax3.set_xlim(0, 8)
ax3.set_ylim(0.74, 1.6)
ax4.set_xlim(0, 8)
ax4.set_ylim(0, 0.33)

ax1.get_xaxis().set_visible(False)
ax2.get_xaxis().set_visible(False)
ax3.get_xaxis().set_visible(False)

ax1.set_ylabel("$\sigma_{xx}/x_{\\rm{zp}}^2$", labelpad=14)
ax2.set_ylabel("$\sigma_{xp}/(x_{\\rm{zp}}p_{\\rm{zp}})$")
ax3.set_ylabel("$\sigma_{pp}/p_{\\rm{zp}}^2$", labelpad=14)
ax3.set_xlabel("$\\Omega \, t$")
ax4.set_ylabel("$\sigma_{xx} \, \\sigma_{pp}$", labelpad=16)
ax4.set_xlabel("$\\Omega \, t$")

ax4.set_yticks([0, 0.25, 0.5])
ax4.set_yticklabels(["0", "$\\frac{\hbar^2}{4}$", "$\\frac{\hbar^2}{2}$"])

n = 5

# ax1.plot(var_t_osys_prm3_cl['t'], 2*var_t_osys_prm3_cl['var_xx'], linestyle='-', linewidth=2, color='skyblue')
ax1.plot(var_t_osys_prm3_cl_mm[:,0], 2*var_t_osys_prm3_cl_mm[:,1], color=lightblue, linestyle='-', linewidth=1.5)
ax1.plot(var_t_sto_prm3_cl['t'][::n], 2*var_t_sto_prm3_cl['var_xx'][::n], color=lightblue, linestyle='', marker='o', markersize=4)
# ax1.plot(var_t_osys_prm3['t'], 2*var_t_osys_prm3['var_xx'], 'C0-', linewidth=2)
ax1.plot(var_t_osys_prm3_mm[:,0], 2*var_t_osys_prm3_mm[:,1], color=blue, linestyle='-', linewidth=1.5)
ax1.plot(var_t_sto_prm3['t'][::n], 2*var_t_sto_prm3['var_xx'][::n], color=blue, linestyle='', marker='o', markersize=4)
ax1.text(0.01, 0.95, '\\textbf{a)}', fontsize = 10, transform = ax1.transAxes, ha='left', va='top') 

# ax2.plot(var_t_osys_prm3_cl['t'], 2*var_t_osys_prm3_cl['var_xp'], linestyle='-', linewidth=2, color='skyblue')
ax2.plot(var_t_osys_prm3_cl_mm[:,0], 2*var_t_osys_prm3_cl_mm[:,2], color=lightblue, linestyle='-', linewidth=1.5)
ax2.plot(var_t_sto_prm3_cl['t'][::n], 2*var_t_sto_prm3_cl['var_xp'][::n], color=lightblue, linestyle='', marker='o', markersize=4)
# ax2.plot(var_t_osys_prm3['t'], 2*var_t_osys_prm3['var_xp'], 'C0-', linewidth=2)
ax2.plot(var_t_osys_prm3_mm[:,0], 2*var_t_osys_prm3_mm[:,2], color=blue, linestyle='-', linewidth=1.5)
ax2.plot(var_t_sto_prm3['t'][::n], 2*var_t_sto_prm3['var_xp'][::n], color=blue, linestyle='', marker='o', markersize=4)
ax2.text(0.01, 0.95, '\\textbf{b)}', fontsize = 10, transform = ax2.transAxes, ha='left', va='top') 

# ax3.plot(var_t_osys_prm3_cl['t'], 2*var_t_osys_prm3_cl['var_pp'], linestyle='-', linewidth=2, color='skyblue')
ax3.plot(var_t_osys_prm3_cl_mm[:,0], 2*var_t_osys_prm3_cl_mm[:,4], color=lightblue, linestyle='-', linewidth=1.5)
ax3.plot(var_t_sto_prm3_cl['t'][::n], 2*var_t_sto_prm3_cl['var_pp'][::n], color=lightblue, linestyle='', marker='o', markersize=4)
# ax3.plot(var_t_osys_prm3['t'], 2*var_t_osys_prm3['var_pp'], 'C0-', linewidth=2)
ax3.plot(var_t_osys_prm3_mm[:,0], 2*var_t_osys_prm3_mm[:,4], color=blue, linestyle='-', linewidth=1.5)
ax3.plot(var_t_sto_prm3['t'][::n], 2*var_t_sto_prm3['var_pp'][::n], color=blue, linestyle='', marker='o', markersize=4)
ax3.text(0.01, 0.95, '\\textbf{c)}', fontsize = 10, transform = ax3.transAxes, ha='left', va='top') 

ax4.fill_between(var_t_sto_prm3['t'], 0, 0.25, color='whitesmoke', zorder=0)
ax4.plot(var_t_osys_prm3_mm[:,0], var_t_osys_prm3_mm[:,1]*var_t_osys_prm3_mm[:,4], color=blue, label = 'OQu', linewidth=1.5, zorder=1)
ax4.plot(var_t_osys_prm3_cl_mm[:,0], var_t_osys_prm3_cl_mm[:,1]*var_t_osys_prm3_cl_mm[:,4], color=lightblue, label = 'OCl', linewidth=1.5, zorder=0)
ax4.axhline(0.25, linestyle='--', color='k')

ax4.text(0.9, 0.85, '$\\sigma_{xx} \sigma_{pp} > \\frac{\hbar^2}{4}$', fontsize = 10, transform = ax4.transAxes, ha='right', va='top') 
ax4.text(0.9, 0.15, '$\\sigma_{xx} \sigma_{pp} < \\frac{\hbar^2}{4}$', fontsize = 10, transform = ax4.transAxes, ha='right', va='bottom')
ax4.text(0.01, 0.97, '\\textbf{d)}', fontsize = 10, transform = ax4.transAxes, ha='left', va='top') 

handles, labels = plt.gca().get_legend_handles_labels()
qu_sto = matplotlib.lines.Line2D([0], [0], label='StoQu', marker='o', markersize=5, markerfacecolor=blue, markeredgecolor='none', linestyle='')
cl_sto = matplotlib.lines.Line2D([0], [0], label='StoCl', marker='o', markersize=5, markerfacecolor=lightblue, markeredgecolor='none', linestyle='')
handles.extend([qu_sto, cl_sto])      
labels.extend(['StoQu', 'StoCl']) 
order = [0, 2, 1, 3]
ax4.legend([handles[idx] for idx in order],[labels[idx] for idx in order], loc='lower left', ncol=1, columnspacing=0.4, labelspacing=0.2, frameon=False)

plt.savefig('./data/dynamics/dynamics.pdf', bbox_inches="tight", dpi=600)
```
